```python
import pandas as pd
import os
```

```python
directory_path = 'data/twitter/'
files = os.listdir(directory_path)
data = []
```

```python
for file in files:
    if file.endswith('.csv'):
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        data.append(df)

combined_data = pd.concat(data)
```
