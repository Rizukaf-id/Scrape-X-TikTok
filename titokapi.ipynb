```python
url_comment = "https://tiktok-api23.p.rapidapi.com/api/post/comments"
url_reply = "https://tiktok-api23.p.rapidapi.com/api/post/comment/replies"
headers = {
    "x-rapidapi-key": "a8c4ed5b7dmsh9810575efeb725fp186d0ajsn0875bf677f8b",
    "x-rapidapi-host": "tiktok-api23.p.rapidapi.com"
}

n = 0
for id in video_id:
    n += 1
    file_path = f'./data/tiktok/batch_4/{id}.csv'
    
    if not os.path.exists(file_path):
        querystring_comment = {"videoId": f"{id}", "count": "2000", "cursor": "0"}

        response_comment = requests.get(url_comment, headers=headers, params=querystring_comment)
        if response_comment.status_code == 200:
            data_comment = response_comment.json()
            if 'comments' in data_comment and data_comment['comments'] is not None:
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write('cid;comment;reply_cid;reply_id;reply_text\n')
                    for comment in data_comment['comments']:
                        cid = comment['cid']
                        text = comment['text']
                        querystring_reply = {"videoId": f"{id}", "commentId": f"{cid}", "count": "1000", "cursor": "0"}
                        response_reply = requests.get(url_reply, headers=headers, params=querystring_reply)
                        data_reply = response_reply.json()
                        if 'comments' in data_reply and data_reply['comments'] is not None:
                            for reply in data_reply['comments']:
                                reply_cid = reply['cid']
                                reply_id = reply['reply_id']
                                reply_text = reply['text']
                                f.write(f'{cid};{text};{reply_cid};{reply_id};{reply_text}\n')
                                print(f'{n}| {cid};{text};{reply_cid};{reply_id};{reply_text}')
                        else:
                            f.write(f'{cid};{text};None;None;None\n')
                            print(f'{n}| {cid},{text}')
            else:
                print(f"{n}| No comments found for video ID {id}")
        else:
            print(f"{n}| Failed to fetch comments for video ID {id}, HTTP status code: {response_comment.status_code}")
    else:
        print(f"{n}| File {file_path} already exists, skipping...")
```

```python
# Define the directory path
directory_path = 'data/tiktok/batch_4'

# Ensure the directory exists to avoid FileNotFoundError
if os.path.exists(directory_path):
    # List all files in the directory
    all_files = os.listdir(directory_path)

    # Filter for CSV files
    csv_files = [file for file in all_files if file.endswith('.csv')]

    # Initialize an empty list to hold dataframes
    dataframes = []

    # Loop through the CSV files, read them into dataframes, and append to the list
    for file in csv_files:
        file_path = os.path.join(directory_path, file)
        try:
            # Attempt to read the CSV file with updated parameters to handle errors
            df = pd.read_csv(file_path, delimiter=';', on_bad_lines='skip')
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")

    # Concatenate all dataframes into a single dataframe, only if there are dataframes to concatenate
    if dataframes:
        combined_df = pd.concat(dataframes, ignore_index=True)

        # Optional: Save the combined dataframe to a new CSV file
        combined_df.to_csv('data/tiktok/combined_data/new_combined_tiktok_data_batch_4.csv', index=False)
else:
    print(f"Directory {directory_path} does not exist.")
```
